```python
import os
import pandas as pd

#MUST HAVE THE RIGHT INPUT DATA. USE Date instead of Term Start, and add the hours to the data for combined datetime 

# Directory path
new_directory = "D:\\Users\\Krawczyk\\Downloads\\Berkshire project\\Untitled Folder "

# Change the current working directory
os.chdir(new_directory)

# Define a path for output directory
out_path = 'D:\\Users\\Krawczyk\\Downloads\\Berkshire project\\Untitled Folder\\NSL files\\Processed'

# Print the current working directory to verify that it has been changed
print(os.getcwd())

# Define the directory containing the CSV files
csv_dir = 'D:\\Users\\Krawczyk\\Downloads\\Berkshire project\\Untitled Folder\\NSL files\\'

# Create an empty list to store the names of the CSV files
csv_files = []

# Loop through all the files in the directory
for filename in os.listdir(csv_dir):
    # Check if the file is a CSV file
    if filename.endswith('.csv'):
        # Add the file name to the list of CSV files
        csv_files.append(filename)
        
def process_csv_file(filename):
    # Load hourly data for 7 years into a DataFrame
    data = pd.read_csv(filename)

    # Replace any occurrences of '#VALUE!' with 'NaN'
    data = data.replace("#VALUE!", float("NaN"))

    # Convert date column to Pandas datetime format
    data["Date"] = pd.to_datetime(data["Date"], format="%m/%d/%Y %H:%M")

    # Add 'Month', 'Day', and 'Hour' columns to the data frame
    data['Month'] = data['Date'].dt.month
    data['Day'] = data['Date'].dt.day
    data['Hour'] = data['Date'].dt.hour

    # Set multi-index for the data frame
    data = data.set_index(['Month', 'Day', 'Hour'])

    # Calculate hourly averages for each hour of the year
    hourly_averages = data.groupby(['Month', 'Day', 'Hour'])['Value'].mean().to_frame()

    # Reset the index to a simple range index
    hourly_averages = hourly_averages.reset_index()

    # Extract the start and end dates from the data frame
    start_date = data['Date'].iloc[0].strftime('%Y-%m-%d')
    end_date = data['Date'].iloc[-1].strftime('%Y-%m-%d')

    # Create output path for the CSV file with start and end dates in the name
    out_file = os.path.join(out_path, f"{os.path.basename(filename)}_{start_date}_{end_date}.csv")

    # Save the resulting data frame to a CSV file in the output directory
    hourly_averages.to_csv(out_file, index=False)

    # Print a message indicating the file has been processed
    print(f"File {filename} processed successfully and saved to {out_file}")


# Loop through the list of CSV files and run the application on each one
for filename in csv_files:
    process_csv_file(os.path.join(csv_dir, filename))
```
